  Graph for the first Distribution.

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the normal distribution
mu = 1  # Mean (μ)
sigma = 1  # Standard deviation (σ)

# Create a range of x values from -5 to 7
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Calculate the normal distribution's probability density function (PDF)
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Plot the normal distribution
plt.figure(figsize=(8, 4))
plt.plot(x, pdf, 'k-', linewidth=2)
plt.title('Normal Distribution (μ=1, σ=1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

Merge the two densities.

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the normal distributions
mu1 = 1  # Mean (μ) for the first distribution
sigma1 = 1  # Standard deviation (σ) for the first distribution

mu2 = -1  # Mean (μ) for the second distribution
sigma2 = 1  # Standard deviation (σ) for the second distribution

# Create a range of x values that cover both distributions
x = np.linspace(min(mu1 - 4*sigma1, mu2 - 4*sigma2), max(mu1 + 4*sigma1, mu2 + 4*sigma2), 100)

# Calculate the normal distribution's probability density function (PDF) for both distributions
pdf1 = (1 / (sigma1 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu1) / sigma1)**2)
pdf2 = (1 / (sigma2 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu2) / sigma2)**2)

# Plot the normal distributions
plt.figure(figsize=(8, 4))
plt.plot(x, pdf1, 'b-', linewidth=2, label='μ=1, σ=1')
plt.plot(x, pdf2, 'r-', linewidth=2, label='μ=-1, σ=1')
plt.title('Comparison of Normal Distributions')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```
